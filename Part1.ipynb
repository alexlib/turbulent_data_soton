```python
U = loadtxt('WindtunnelSample1.txt')

acq_freq = 60000  # sampling frequency is 60kHz
dt = 1./acq_freq  # time interval between successive data points
t = arange(0,(U.shape[0])*dt,dt)  # time of each sample
```

```python
# plot the data, first 1000 points
plot(t,U);
xlabel('$t$ (sec)')
ylabel('$U$ (m/s)');
```

```python
# calculate the mean
Um = mean(U)
```

```python
Um
```

```python
# new fluctuating velocity
u = U - Um
```

```python
plot(t,u)
xlabel('$t$ (sec)')
ylabel('$u$ (m/sec)');
```

```python
# variance, skewness, kurtosis
variance_u = sum(u**2)/u.shape[0]
skewness_u = (sum(u**3)/u.shape[0]) / (variance_u**(3/2))
kurtosis_u = (sum(u**4)/u.shape[0]) / (variance_u**(2))

print(variance_u, skewness_u, kurtosis_u)
```

```python
ax = arange(-6,6,.01) # [-6:.01:6]; %set up the bins for the pdf
ay = histogram(u, bins=ax) # calculate the histogram of u
```

```python
plot(ax[:-1],ay[0]/trapz(ay[0],ax[:-1]))
```

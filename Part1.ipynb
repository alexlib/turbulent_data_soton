```python
import os, glob
```

```python
glob.glob('*.txt')
```

```python
d = loadtxt('WindtunnelSample1.txt')
```

```python
# plot the data, first 1000 points
plot(d[:1000])
```

```python
# calculate the mean
U = mean(d)
```

```python
U
```

```python
# new fluctuating velocity
uf = d - U
```

```python
plot(uf)
```

```python
# variance, skewness, kurtosis
var(uf)
```

```python
from scipy.stats import skew, kurtosis
```

```python
skew(uf)
```

```python
kurtosis(uf)
```

```python
ax = arange(-6,6,.01) # [-6:.01:6]; %set up the bins for the pdf
ay = histogram(uf, bins=ax) # calculate the histogram of u
```

```python
plot(ax[:-1],ay[0]/trapz(ay[0],ax[:-1]))
```
